MITgcm sensitivity to prescribed values of vertical and horizontal diffusivity
-----------------------------------------------------------------------

The following values were computed from MITgcm output for runs: run01notave, run02notave,run03notave,run04notave, in experiment 360x360x90_BodyForcing_6Tr_LinProfiles. The calculations were performes in the ipython notebooks NumDiffSensitivityTestnoTaveRun0#.ipynb with #=1,2,3,4. They are stored in my GDrive file DiffusivityExperimentsCalculations.

```python
Kh1=[1.00E-07,1.00E-05,1.00E-02,1,20,50]
Kv1=[1.00E-05,1.00E-05,1.00E-05,1.00E-05,1.00E-05,1.00E-05]

Kh2=[1.00E-05,1.00E-05,1.00E-05,1.00E-05,1.00E-05,1.00E-05]
Kv2=[1.00E-07,1.00E-03,1.00E-02,1.00E-01,1,10]

Khnum1=[-0.16499406,-0.16499406,-0.16499451,-0.16502153,-0.16522534,-0.16524882]
Khnum1sd=[8.00E-03,8.00E-03,8.00E-03,8.00E-03,8.00E-03,8.00E-03]
Kvnum1=[-3.38E-05,-3.38E-05,-3.38E-05,-3.39E-05,-3.45E-05,-3.48E-05]
Kvnum1sd=[1.62E-06,1.62E-06,1.62E-06,1.63E-06,1.72E-06,1.76E-06]

Khnum2=[-0.1661,-0.0666,0.7139,6.2184,18.9489,11.402]
Khnum2sd=[0.008,0.008,0.016,0.335,5.368,9.49]
Kvnum2=[-3.39E-05,-1.44E-05,0.0002,0.0015,0.0024,0.0011]
Kvnum2sd=[1.61E-06,1.82E-06,3.40E-06,9.43E-05,0.001,0.001]

Khnum3=[0.9161,0.9161,0.933,2.1657,9.8988,13.524]
Khnum3sd=[0.08,0.08,0.08,0.064,0.545,0.8]
Kvnum3=[0.0002,0.0002,0.0002,0.0002,0.0005,0.0006]
Kvnum3sd=[7.95E-06,7.95E-06,8.07E-06,1.58E-05,4.53E-05,4.52E-05]

Khnum4=[0.9041,1.3797,2.1681,3.3772,4.198,2.6962]
Khnum4sd=[0.08,0.096,0.159,0.23,0.283,0.253]
Kvnum4=[0.0002,0.0002,0.0003,0.0003,0.0003,0.0003]
Kvnum4sd=[7.52E-06,1.37E-05,2.04E-05,2.90E-05,3.25E-05,2.42E-05]
```

Description of the runs:

Upwelling-favorable body forcing, 10 days (21600 timesteps), output every 12 hours, Smagorinsky (2.2), ran on Bugaboo.
Run01notave and run02notave have a vertical initial tracer profile thay decays linearly with depth. In run01notave Kv is kept constant (1E-5 $m^2s^{-1}$) while Kh varies from 1E-7 to 50 $m^2s^{-1}$; in run02notave Kh is kept constant (1E-5 $m^2s^{-1}$) while Kv varies from 1E-7 to 10 $m^2s^{-1}$.

Run03notave and run04notave have a horizontal initial tracer profile thay decays linearly away from the shelf. In run03notave Kv is kept constant (1E-5 $m^2s^{-1}$) while Kh varies from 1E-7 to 50 $m^2s^{-1}$; in run04notave Kh is kept constant (1E-5 $m^2s^{-1}$) while Kv varies from 1E-7 to 10 $m^2s^{-1}$.

Two additional runs, run05notave and run06notave, were executed to add more data points to the vertical diffusivity variation plot. run05notave adds values of Kv 1E-5 and 1E-4 $m^2s^{-1}$ to run02notave and run06notave adds values of Kv 1E-5 and 1E-4 $m^2s^{-1}$ to run04notave.

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 15}

plt.rc('font', **font)
rcParams['legend.numpoints'] = 1


fig, ax1 = plt.subplots(figsize=(10,8))


lns1=ax1.errorbar(Kh1,Khnum1, fmt=':b^',label='run01 vgrad')
lns2=ax1.errorbar(Kh1,Khnum3,yerr=Khnum3sd,fmt= ':bs',label='run03 hgrad')
ax1.set_xlabel('Kh implicit ($m^2s^{-1}$)')
ax1.set_xscale('log')
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Kh num ($m^2s^{-1}$)', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')


ax2 = ax1.twinx()

lns3=ax2.errorbar(Kh1,Kvnum1 , fmt=':r^',label='run01 vgrad')
lns4=ax2.errorbar(Kh1,Kvnum3 ,yerr=Kvnum3sd,fmt= ':rs',label='run03 hgrad')
ax2.set_ylabel('Kv num ($m^2s^{-1}$)', color='r')

for tl in ax2.get_yticklabels():
    tl.set_color('r')

ax1.legend(loc=6)
ax2.legend(loc=0)

#lns = lns1+lns2+lns3+lns4
#labs = [l.get_label() for l in lns]
#ax1.legend(lns, labs, loc=0)
    
plt.title('Constant implicit Kv (1E-5 $m^2s^{-1}$)')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,8))


lns1=ax1.errorbar(Kv2,Khnum2,yerr=Khnum2sd,  fmt=':b^',label='run02 vgrad')
lns2=ax1.errorbar(Kv2,Khnum4,yerr=Khnum4sd, fmt=':bs',label='run04 hgrad')

ax1.set_xlabel('Kv implicit ($m^2s^{-1}$)')
ax1.set_xscale('log')

# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Kh num ($m^2s^{-1}$)', color='b')

for tl in ax1.get_yticklabels():
    tl.set_color('b')


ax2 = ax1.twinx()

lns3=ax2.errorbar(Kv2,Kvnum2 ,yerr=Kvnum2sd,fmt=':r^',label='run02 vgrad')
lns4=ax2.errorbar(Kv2,Kvnum4 ,yerr=Kvnum4sd,fmt=':rs',label='run04 hgrad')

ax2.set_ylabel('Kv num ($m^2s^{-1}$)', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

#lns = lns1+lns2+lns3+lns4
#labs = [l.get_label() for l in lns]
#ax1.legend(lns, labs, loc=0)
ax1.legend(loc=6)
ax2.legend(loc=0)
#ax1.xlim((1e-8,20))
#ax2.xlim((1e-8,20))

plt.title('Constant implicit Kh (1E-5 $m^2s^{-1}$)')

```
