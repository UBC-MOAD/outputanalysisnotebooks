## Location and definition of transport cross-sections

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
import xarray as xr
```

```python
grid_file = '/ocean/kramosmu/MITgcm/TracerExperiments/CNTDIFF/run38/gridGlob.nc'
grid = xr.open_dataset(grid_file)
```

```python
CS1_x_slice = slice(0, 59)
CS2_x_slice = slice(60, 119)
CS3_x_slice = slice(120, 239)
CS4_x_slice = slice(240, 299)
CS5_x_slice = slice(300, 359)

CS1_y_ind = 227
CS1b_y_ind = 267
CS1c_y_ind = 297
CS1d_y_ind = 359

AS1_y_slice = slice(227, 266)
AS2_y_slice = slice(267, 298)
AS3_y_slice = slice(299, 359)

AS1_x_ind = 59
AS1b_x_ind = 119
AS1c_x_ind = 239
AS1d_x_ind = 299
```

```python
sns.set_style(style='white')
sns.set_context('talk')
fig, ax = plt.subplots(1, 1, figsize=(8,6))

ax.contour(grid.X/1000,grid.Y/1000,grid.Depth,[50,100,150,200,400,600,800,1000], colors='0.4')

ax.plot(grid.XC[CS1_y_ind,CS1_x_slice]/1000,grid.YC[CS1_y_ind,CS1_x_slice]/1000, 'b-', linewidth=3)
ax.plot(grid.XC[CS1_y_ind,CS2_x_slice]/1000,grid.YC[CS1_y_ind,CS2_x_slice]/1000, 'g-', linewidth=3)
ax.plot(grid.XC[CS1_y_ind,CS3_x_slice]/1000,grid.YC[CS1_y_ind,CS3_x_slice]/1000, 'r-', linewidth=3)
ax.plot(grid.XC[CS1_y_ind,CS4_x_slice]/1000,grid.YC[CS1_y_ind,CS4_x_slice]/1000, '-',color='purple', linewidth=3)
ax.plot(grid.XC[CS1_y_ind,CS5_x_slice]/1000,grid.YC[CS1_y_ind,CS5_x_slice]/1000, '-', color='orange', linewidth=3)

ax.plot(grid.XC[CS1b_y_ind,CS1_x_slice]/1000,grid.YC[CS1b_y_ind,CS1_x_slice]/1000, 'b-', linewidth=3)
ax.plot(grid.XC[CS1b_y_ind,CS2_x_slice]/1000,grid.YC[CS1b_y_ind,CS2_x_slice]/1000, 'g-', linewidth=3)
ax.plot(grid.XC[CS1b_y_ind,CS3_x_slice]/1000,grid.YC[CS1b_y_ind,CS3_x_slice]/1000, 'r-', linewidth=3)
ax.plot(grid.XC[CS1b_y_ind,CS4_x_slice]/1000,grid.YC[CS1b_y_ind,CS4_x_slice]/1000, '-',color='purple', linewidth=3)
ax.plot(grid.XC[CS1b_y_ind,CS5_x_slice]/1000,grid.YC[CS1b_y_ind,CS5_x_slice]/1000, '-', color='orange', linewidth=3)

ax.plot(grid.XC[CS1c_y_ind,CS1_x_slice]/1000,grid.YC[CS1c_y_ind,CS1_x_slice]/1000, 'b-', linewidth=3)
ax.plot(grid.XC[CS1c_y_ind,CS2_x_slice]/1000,grid.YC[CS1c_y_ind,CS2_x_slice]/1000, 'g-', linewidth=3)
ax.plot(grid.XC[CS1c_y_ind,CS3_x_slice]/1000,grid.YC[CS1c_y_ind,CS3_x_slice]/1000, 'r-', linewidth=3)
ax.plot(grid.XC[CS1c_y_ind,CS4_x_slice]/1000,grid.YC[CS1c_y_ind,CS4_x_slice]/1000, '-',color='purple', linewidth=3)
ax.plot(grid.XC[CS1c_y_ind,CS5_x_slice]/1000,grid.YC[CS1c_y_ind,CS5_x_slice]/1000, '-', color='orange', linewidth=3)

ax.plot(grid.XC[AS1_y_slice,AS1_x_ind]/1000,grid.YC[AS1_y_slice,AS1_x_ind]/1000, '-', color='0.2',linewidth=3)
ax.plot(grid.XC[AS2_y_slice,AS1_x_ind]/1000,grid.YC[AS2_y_slice,AS1_x_ind]/1000, '-', color='0.5',linewidth=3)
ax.plot(grid.XC[AS3_y_slice,AS1_x_ind]/1000,grid.YC[AS3_y_slice,AS1_x_ind]/1000, '-', color='0.8',linewidth=3)

ax.plot(grid.XC[AS1_y_slice,AS1b_x_ind]/1000,grid.YC[AS1_y_slice,AS1b_x_ind]/1000, '-', color='0.2', linewidth=3)
ax.plot(grid.XC[AS2_y_slice,AS1b_x_ind]/1000,grid.YC[AS2_y_slice,AS1b_x_ind]/1000, '-', color='0.5', linewidth=3)
ax.plot(grid.XC[AS3_y_slice,AS1b_x_ind]/1000,grid.YC[AS3_y_slice,AS1b_x_ind]/1000, '-', color='0.8', linewidth=3)

ax.plot(grid.XC[AS1_y_slice,AS1c_x_ind]/1000,grid.YC[AS1_y_slice,AS1c_x_ind]/1000, '-', color='0.2', linewidth=3)
ax.plot(grid.XC[AS2_y_slice,AS1c_x_ind]/1000,grid.YC[AS2_y_slice,AS1c_x_ind]/1000, '-', color='0.5', linewidth=3)
ax.plot(grid.XC[AS3_y_slice,AS1c_x_ind]/1000,grid.YC[AS3_y_slice,AS1c_x_ind]/1000, '-', color='0.8', linewidth=3)

ax.plot(grid.XC[AS1_y_slice,AS1d_x_ind]/1000,grid.YC[AS1_y_slice,AS1d_x_ind]/1000, '-', color='0.2', linewidth=3)
ax.plot(grid.XC[AS2_y_slice,AS1d_x_ind]/1000,grid.YC[AS2_y_slice,AS1d_x_ind]/1000, '-', color='0.5', linewidth=3)
ax.plot(grid.XC[AS3_y_slice,AS1d_x_ind]/1000,grid.YC[AS3_y_slice,AS1d_x_ind]/1000, '-', color='0.8', linewidth=3)


ax.set_xlabel('alongshore distance [km]')
ax.set_ylabel('cross-shore distance [km]')

```
