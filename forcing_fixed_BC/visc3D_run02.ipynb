# 3D vertical viscosity and vertical diffusivity- run02

First complete run with 3D vertical viscosity. I modified calc_viscosity.F to call calc_3Ddiffusivity.F and set Av=Kv. This run takes the 3D kv file with $10^{-5}m^2s^{-1}$ outside the canyon and $10^{-3} m^2s^{-1}$ within the canyon. It ran for 9 days.


```python
from math import *
import matplotlib.pyplot as plt
%matplotlib inline
import os
from netCDF4 import Dataset
import numpy as np
import pandas as pd
import seaborn as sns
import struct
import xarray as xr
```
